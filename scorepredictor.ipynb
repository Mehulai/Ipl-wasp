```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("ipldataset2.csv")
df.head()
```

```python
data = df[['Team1','Team2','Score1','Score2','venue']]
data.head()
```

```python
c = ['KKR','RCB','CSK','MI','DD','SRH','RR','KXIP']
d = ['KKR','RCB','CSK','MI','DD','SRH','RR','KXIP']
e = ['Kolkata','Bangalore','Chennai','Mumbai','Delhi','Hyderabad','Jaipur','Mohali']
for j in range(0,len(c)):
    a = []
    b = []
    for k in range(0,len(d)):
        for i in range(0,len(data)):
            if(j==k):
                break
            if(data.iloc[i][0]== c[j] and data.iloc[i][1]==d[k] and data.iloc[i][4]==e[j]):
                a.append(data.iloc[i][2])
                b.append(data.iloc[i][3])
        if(j==k || a==[] || b == []):
            pass
        else:
            print(c[j],'=',round(sum(a)/len(a)))
            print(d[k],'=',round(sum(b)/len(b)))
            print('-------')
```

```python
c='DD'
d='RCB'
e='Delhi'
for i in range(0,len(data)):
    if(data.iloc[i][0]==c and data.iloc[i][1]==d and data.iloc[i][4]==e):
         a.append(data.iloc[i][2])
         b.append(data.iloc[i][3])
print(c,'=',round(sum(a)/len(a)))
print(d,'=',round(sum(b)/len(b)))
```
