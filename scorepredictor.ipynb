```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("ipldataset2.csv")
df.head()
```

```python
data = df[['Team1','Team2','Score1','Score2','venue']]
data.head()
```

```python
a = []
b= []
c = 'KKR'
d = 'RCB'
e = 'Kolkata'
for i in range(0,len(data)):
    if(data.iloc[i][0]== c and data.iloc[i][1]==d and data.iloc[i][4] == e):
        a.append(data.iloc[i][2])
        b.append(data.iloc[i][3])
print(c,'=',round(sum(a)/len(a)))
print(d,'=',round(sum(b)/len(b)))
```
